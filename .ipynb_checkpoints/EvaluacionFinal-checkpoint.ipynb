## 1. Realizar un análisis exploratorio de los datos para identificar relaciones entre variables, valores atípicos, tendencias, etc.

```python
#cargando librerias

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from  sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.metrics import accuracy_score
```

```python
Data = pd.read_csv('Titanic-Dataset.csv')
```

```python
Data.head(50)
```

```python
Data.describe()
```

Convenciones Datos

###### PassengerId : Passenger ID
###### Survived: Weather Survived or not: 0 = No, 1 = Yes
###### Pclass: Ticket class: 1 = 1st, 2 = 2nd, 3 = 3rd
###### Name: Name of the Passenger
###### Sex: Gender
###### Age: Age in Years
###### SibSp: No. of siblings/spouses aboard the Titanic
###### Parch: No. of parents/children aboard the Titanic

###### Fare: Passenger fare

```python
#identificacion de datos atipicos
plt.figure(figsize=(6,3))
sns.boxplot(x=Data['Age'])
plt.title("Age", fontsize = 10)
```

```python
#identificacion de datos atipicos
plt.figure(figsize=(6,3))
sns.boxplot(x=Data['Pclass'])
plt.title("Pclass", fontsize = 10)
```

## 2. Preprocesar los datos limpiándolos, tratando valores faltantes y transformándolos según sea necesario

```python
Data.isna().sum()
```

```python
Data = Data.drop('Cabin',axis=1)
Data.head()
```

```python
Data = Data.dropna(axis=0)
```

```python
Data.isna().sum()
```

```python

Data.info()
```

```python
Data['Sex'].replace({'male':0, 'female':1}, inplace=True)
```

```python
Data.head(20)
```

C = Cherbourg = 0, Q = Queenstown = 1, S = Southampton = 2

```python
Data['Embarked'].replace({'C':0, 'Q':1, 'S':2}, inplace=True) 
```

## 3.Seleccionar las características más relevantes para entrenar el modelo utilizando selección de características

```python
Data = Data.drop(['Name','Ticket','Fare','SibSp','Parch',],axis=1)
Data.head()
```



Característica de interes:

    Survived: determina si el pasajero sobrevivio (0 = no sobrevivio,  1 = si sobrevivio)



```python
Data['Survived'].value_counts()
```

```python
Data[['Survived']].head(20)
```

## 4. Dividir el dataset en Train y Test para evaluar correctamente el modelo

```python
# Porcentaje de datos para entrenamiento (train) = 75% , y para prueba (test) = 25%

# Se crean variables X y Y para almacenar la información del entrenamiento

X = Data.drop('Survived',axis=1) 
Y = Data['Survived'] 

#La función train_test_split crea una división de un conjunto de datos en dos bloques uno de entrenamiento y otro de prueba (train and test)
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.25, stratify=Y, random_state= 0)
```

```python
print (X_train)
```

```python
print (Y_train)
```

## 5. Entrenar el modelo configurando los diferentes hiperparámetros

```python
# Construir el modelo de regresión logística 

modelo_regresion = LogisticRegression(solver = "liblinear") 
modelo_regresion.fit (X_train , Y_train)   

# Crea el modelo de prediccion
Prediccion = modelo_regresion.predict(X_test) 
```

```python
print (Y_test)
```

```python
print (Prediccion)
```

## 6. Evaluar el desempeño del modelo en el conjunto de Test con métricas como precisión, recall, F1-score, etc

```python
print(classification_report(Y_test,Prediccion))
```

```python
accuracy = accuracy_score(Y_test,Prediccion)
accuracy
```

## 7. Realizar las diferentes gráficas que permitan visualizar los resultados del modelo

```python
#Matrix de Confusión del modelo

cm=confusion_matrix(Y_test,Prediccion)
disp=metrics.ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot()
```

```python
# Curva de precisón del modelo

from sklearn.metrics import RocCurveDisplay
RocCurveDisplay.from_predictions(Y_test, Prediccion)
plt.show()
```

## 8. Interpretar, analizar y documentar los resultados obtenidos



Del desempeño del modelo se puede describir:

    Precisión: Precisión mide la proporción de predicciones positivas verdaderas entre todas las predicciones positivas realizadas por el modelo. Aquí la precisión para la clase "Survived" es 0,85, lo que significa que el 85% de los caos predichos por el modelo donde sobrevive el pasajero concuerdan con la informacion real
    
    Recall: Mide la proporción de predicciones positivas verdaderas entre todas las instancias positivas reales en el conjunto de datos. Aquí el Recall para la clase "Survived" es 0,83, lo que indica que el modelo identificó correctamente al 83% de todos los casos donde sobrevive el pasajero.

    F1 Score: Es la media armónica de precisión y recuperación. Proporciona una puntuación única que equilibra la Precisión y Recall. Aquí la puntuación F1 para la clase "Survived" es 0,84.

De la Matriz de Confusión podemos interpretar:

    Verdaderos negativos: el recuento de observaciones que un clasificador predijo correctamente como falso 98

    Verdaderos positivos: el recuento de observaciones que un clasificador predijo correctamente como Verdadero 53

    Falsos positivos: el recuento de observaciones que un clasificador predijo incorrectamente como Verdaderas 8

    Falsos negativos: el recuento de observaciones que un clasificador predijo incorrectamente como falso 19

De la curva de precisión del modelo se puede interpretar:

    El AUC es 0,83 lo que significa que el 83 % de las predicciones del clasificador son correctas en todos los umbrales de clasificación.
